```python
from zumi.zumi import Zumi
import matplotlib.pyplot as plt
import numpy as np
import time

speed = 40
time_out = 2

zumi = Zumi()
print(zumi.get_battery_voltage()," volts")
for i in range(127):
    print(i)
    zumi.control_motors(-1*i,-1*i)
    
for i in range(127):
    print(i)
    zumi.control_motors(i,i)

time.sleep(2)
zumi.control_motors(0,0)
zumi.stop()
```

```python

zumi.reset_gyro()

try:
    init_time = time.time()
    time_list = []
    gyro_list = []
    time_passed = 0

    while(True):
        zumi.control_motors(0,speed)
        
        time_passed = time.time()-init_time
        current_angle = zumi.update_angles()[2]
        time_list.append(time_passed)
        gyro_list.append(current_angle) 

        if time_passed >= time_out:
            print(current_angle/time_passed)
            break

        
finally:
    zumi.hard_brake()
    plt.plot(time_list,gyro_list,'ro')
    plt.ylabel('gyro')
    plt.xlabel('time')
    plt.show()
    
```

```python
zumi.reset_gyro()
time_out = 3
try:
    init_time = time.time()
    time_list = []
    gyro_list = []
    batt_list = []
    time_passed = 0

    while(True):
        
        time_passed = time.time()-init_time
        current_angle = zumi.update_angles()[2]
        time_list.append(time_passed)
        gyro_list.append(current_angle)
        battery_level = zumi.get_battery_voltage()
        batt_list.append(battery_level)
        #zumi.control_motors(-speed,-speed)
        #zumi.go_straight(40,0)
        accuracy = 0
        zumi.drive_at_angle(127, 40, 0, 3, 0.3, 0.01, accuracy)

        if time_passed >= time_out:
            print(current_angle/time_passed)
            break
      
finally:
    zumi.hard_brake()
    plt.plot(time_list,gyro_list,'go',time_list,batt_list,'bo')
    plt.ylabel('gyro')
    plt.xlabel('time')
    plt.show()
```
