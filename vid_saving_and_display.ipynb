```python
from zumi.util.camera import Camera
import PIL.Image
import cv2
import IPython.display
import time

width = 160*4
height = 120*4

camera = Camera(width,height)
camera.start_camera()

try:
    fourcc = cv2.VideoWriter_fourcc(*'H264')
    #fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    #fourcc = cv2.VideoWriter_fourcc(*'X264')
    #fourcc = cv2.VideoWriter_fourcc(*'XVID')
    #fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    #video_writer = cv2.VideoWriter('output.mp4',fourcc, 30.0, (width, height))
    #video_writer = cv2.VideoWriter('output.avi',fourcc, 20.0, (width, height))
    video_writer = cv2.VideoWriter('output.mp4',fourcc, 5.0, (width, height))
    
    # record video
    for i in range(20):
        time_start = time.time()
        print(i)
        image = camera.capture()
        #resized = cv2.resize(image, (40*2, 30*2))
        #IPython.display.display(PIL.Image.fromarray(resized))
        image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        video_writer.write(image)
        print(time.time()-time_start)
        !free -h
        #IPython.display.clear_output(wait=True)
finally:
    time.sleep(1)
    camera.close()
    video_writer.release()
    print("done")
    
IPython.display.Video("output.mp4")
```

```python
from zumi.util.camera import Camera
import cv2
import IPython.display
import PIL.Image
from io import BytesIO

camera = Camera(1*80,1*60)
camera.start_camera()

try: 

    while True:
        frame = camera.capture()
        #IPython.display.display(PIL.Image.fromarray(frame))
        img = PIL.Image.fromarray(frame, "RGB")
        buffer = BytesIO()
        img.save(buffer,format="JPEG") 
        IPython.display.display(IPython.display.Image(data=buffer.getvalue()))


        IPython.display.clear_output(wait=True) 

finally:
    camera.close()
```
