```python
from zumi.zumi import Zumi
from zumi.util.screen import Screen
import matplotlib.pyplot as plt
import numpy as np
import time

zumi = Zumi()
zumi.mpu.calibrate_MPU(100)
zumi.reset_gyro()
time_list = []
ang_z_list = []
ang_z_speed_list = []

time_out = 10
time_passed = 0
init_time = time.time()  
print("start")       
while(True):
    time_passed = time.time()-init_time
    #add data to lists
    time_list.append(time_passed)
    zumi.update_angles()
    #zumi.angle_list[2] = zumi.angle_list[2] + 0.00893
    #record the z angle
    ang_z_list.append(zumi.angle_list[2])
    ang_z_speed_list.append(zumi.mpu_list[5])

    #program should end if timeout or number of bricks
    if time_passed > time_out:
        break
print("done")  
print(len(time_list))
zumi.hard_brake()


plt.plot(time_list,ang_z_speed_list,'ro')
plt.ylabel('gyro angular speed')
plt.xlabel('time')
plt.show()

plt.plot(time_list,ang_z_list,'bo')
plt.ylabel('gyro angle')
plt.xlabel('time')
plt.show()

m, b = np.polyfit(time_list, ang_z_list, 1)
print("pred deg ", m)
print("slope int ",b)

# m, b = np.polyfit(time_list, ang_z_speed_list, 1)
# print("pred deg/s ", m)
# print("slope int ",b)
```
