```python
from zumi.util.camera import Camera
from zumi.util.screen import Screen
from zumi.zumi import Zumi
import cv2
import IPython.display
import PIL.Image
import time
import math
import numpy as np
from scipy.stats import mode

screen = Screen()
zumi = Zumi()

#width = 1280
#height = 960

#width = 640
#height = 480

#width = 320
#height = 240

width = 160 
height = 120

#width = 80
#height = 60

black_white_threshold = 100


def track_this_hue(image,color):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    filteredFrame = cv2.inRange(hsv, color[0], color[1])
    colorCutout =  cv2.bitwise_and(image, image, mask=filteredFrame)

    return colorCutout,filteredFrame

#given a list of lines
def add_lines(lines,image,ang_thresh=45):

    if lines is not None:
        lst = []
        cnt_x_points = []
        cnt_y_points = []
        for i in range(len(lines)):
            for x1,y1,x2,y2 in lines[i-1]: 
                radians = math.atan2((x2-x1),(y2-y1))
                degrees = math.degrees(radians)
                #filter the angles between 70 and 110 degrees
                #if degrees > -ang_thresh and degrees < ang_thresh:
                if True:
                    lst.append(degrees)
                    cv2.line(image,(x1,y1),(x2,y2),(255,255,255),1)
                    center_x_point = (x2-x1)/2
                    center_y_point = (y2-y1)/2
                    cnt_x_points.append(center_x_point)
                    cnt_y_points.append(center_y_point)                   
```

```python
crop_area = 1.5

try:
    camera = Camera(width,height)
    camera.start_camera()
    while True:
        
        # Take a picture
        image = camera.capture()  
        
        # focus only on the bottom half and extra of the image
        image = image[int(height/crop_area):int(height/crop_area)+height, 0:width]
        
        # Convert it to gray
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

        ret, black_white = cv2.threshold(gray,120,255,cv2.THRESH_BINARY)
        contoursArray = cv2.findContours(black_white.copy(), cv2.RETR_EXTERNAL,
                            cv2.CHAIN_APPROX_SIMPLE)[-2]
        if len(contoursArray) > 0:
            color = 0
            points = []
            
            for contour in contoursArray:
            #contour = contoursArray[-1]
                color = color + 50
                
                x, y, w, h = cv2.boundingRect(contour)
                a = w*h
                if a > 100 and a < 1000 and w > 40 and h < 30:
                    #if the area is big enough and the width is long enough
                    #also since the rectangle have a fixed height we can look for narrow rectangles
                    rectangleColor = (color, 255, 255-color)#draw the rectangle with varying color
                    cv2.rectangle(image, (x, y), (x + w, y + h), rectangleColor, 2)
                    c_x = int(x+w/2)
                    c_y = int(y+h/2)
                    points.append((c_x,c_y))
                    #draw red dot on the center
                    cv2.rectangle(image, (c_x, c_y), (c_x , c_y), (255,0,0), 3)
                    
                    #draw red dot on the left and right of the lanes
                    cv2.rectangle(image, (x, c_y), (x , c_y), (255,0,0), 3)
                    cv2.rectangle(image, (x+w, c_y), (x+w , c_y), (255,0,0), 3)

            if len(points)>1:
                cv2.line(image,points[0],points[1],(255,0,0),3)
                x1,y1 = points[0]
                x2,y2 = points[1]
                radians = math.atan2((x2-x1),(y2-y1))
                degrees = math.degrees(radians)
                #print(degrees)
                
                cv2.putText(image, "ang:" + format(int(degrees)), (10, 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.3, (255,255,255), 1, cv2.LINE_AA) 
                    
        IPython.display.display(PIL.Image.fromarray(image))

        IPython.display.clear_output(wait=True)
        
finally:
    camera.close()
    zumi.stop()
```
