```python
import PIL.Image
import cv2
import IPython.display
import time
from zumi.zumi import Zumi
from zumi.util.screen import Screen
import copy

import math
import time
import random

zumi = Zumi()
screen = Screen()

width = 128
height = 64
center_x = width/2
width_square = 59
x_coordinate = width/2-width_square/2
diameter_dot1 = 15
y_coordinate_dot1 = height/2-diameter_dot1/2
x_coordinate_dot1 = width/2-diameter_dot1/2
screen.draw_square(x_coordinate,2,width_square,fill_in=True)
screen.draw_circle(x_coordinate_dot1, y_coordinate_dot1, diameter_dot1)
#we will display the image on jupyter as well
IPython.display.display(screen.screen_image)
IPython.display.clear_output(wait=True)

def dice_face_1(width_square = 59,diameter_dot1 = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1, y_coordinate_dot1, diameter_dot1)
    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True)
    return copy.deepcopy(screen.screen_image)
    
def dice_face_2(width_square = 59,diameter_dot1 = 12,shift = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1+shift, diameter_dot1)
    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True)
    return copy.deepcopy(screen.screen_image)
    
def dice_face_3(width_square = 59,diameter_dot1 = 12,shift = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1+shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1, y_coordinate_dot1, diameter_dot1)

    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True) 
    return copy.deepcopy(screen.screen_image)

def dice_face_4(width_square = 59,diameter_dot1 = 12,shift = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1+shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1+shift, diameter_dot1)
    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True)
    return copy.deepcopy(screen.screen_image)
 
def dice_face_5(width_square = 59,diameter_dot1 = 12,shift = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1+shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1, y_coordinate_dot1, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1+shift, diameter_dot1)

    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True)
    return copy.deepcopy(screen.screen_image)
    
def dice_face_6(width_square = 59,diameter_dot1 = 12,shift = 15):
    center_x = width/2
    x_coordinate = width/2-width_square/2
    y_coordinate_dot1 = height/2-diameter_dot1/2
    x_coordinate_dot1 = width/2-diameter_dot1/2
    screen.draw_square(x_coordinate,2,width_square,fill_in=True)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1+shift, diameter_dot1)

    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1, diameter_dot1)

    screen.draw_circle(x_coordinate_dot1+shift, y_coordinate_dot1-shift, diameter_dot1)
    screen.draw_circle(x_coordinate_dot1-shift, y_coordinate_dot1+shift, diameter_dot1)

    #we will display the image on jupyter as well
    IPython.display.display(screen.screen_image)
    IPython.display.clear_output(wait=True) 
    return copy.deepcopy(screen.screen_image)


def draw_dice_face(face_img):
    screen.disp.image(face_img)
    screen.disp.display()
    
def roll_dice(number_of_rolls=5):
    for i in range(number_of_rolls):
        val = random.randint(1, 6)
        if val == 1:
            draw_dice_face(d1)
        elif val ==2:
            draw_dice_face(d2)
        elif val ==3:
            draw_dice_face(d3)
        elif val ==4:
            draw_dice_face(d4)
        elif val ==5:
            draw_dice_face(d5)
        elif val ==6:
            draw_dice_face(d6)
        else:
            pass
    time.sleep(1)
    return val

#save each dice face
d1 = dice_face_1()
d2 = dice_face_2()
d3 = dice_face_3()
d4 = dice_face_4()
d5 = dice_face_5()
d6 = dice_face_6()

#this will be the threshold which zumi will detect a tap
#using the accelerometer
tap_threshold = 3


```

```python
#take a couple of samples
while(True):
    data = zumi.update_angles()
    x_acc = data[3]
    if(zumi.get_orientation_message()== "upright"):
        if(abs(x_acc) > tap_threshold):
            print("tapped ")
            #zumi will buzz when tapped
            zumi.play_note(60,30)
            roll_dice()
            
```
