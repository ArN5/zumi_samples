```python
from zumi.util.camera import Camera
import cv2
import IPython.display
import PIL.Image
from time import sleep

stop_sign_path = "stop.xml"
#YOU NEED ^^^ THIS FILE ^^^ DOWNLOAD IT

#credit to https://github.com/cfizette/road-sign-cascades
cascade = cv2.CascadeClassifier(stop_sign_path)
camera = Camera()
camera.start_camera()

#here we will do a single image
frame = camera.capture()
camera.close()
#convert to grayscale
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#find some signs
signs = cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=8, minSize=(40,40), flags=cv2.CASCADE_SCALE_IMAGE)


for (x,y,w,h) in signs:
    #draw a bounding box on the stop sign found
    #x is x positon in frame, y is 
    #w is width of area enclosing stop sign and h is height
    color_of_box = (0,0,255) # this is in (R,G,B) 8 bit 2^8=256
    cv2.rectangle(frame, (x,y), (x+w, y+h), color_of_box, 2)
IPython.display.display(PIL.Image.fromarray(frame))
```

```python
#now lets try video stream of the stop sign
try:
    camera = Camera()
    camera.start_camera()
    while(1):
        frame = camera.capture()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        signs = cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=8, minSize=(40,40), flags=cv2.CASCADE_SCALE_IMAGE)

        for (x,y,w,h) in signs:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)
        #display the image
        IPython.display.display(PIL.Image.fromarray(frame))
        #clear the output for the next frame
        IPython.display.clear_output(wait=True)
        
finally:
    camera.close()
```

```python
#now lets try streaming detecting and showing 
#some text of the location of the stop sign

try:
    camera = Camera()
    camera.start_camera()
    while(1):
        frame = camera.capture()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #you can change the minimum size detected as well. In pixels.
        signs = cascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=8, minSize=(10,10), flags=cv2.CASCADE_SCALE_IMAGE)
        
        for (x,y,w,h) in signs:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)
            #draw the x coordinate on the frame
            cv2.putText(frame, "x:" + format(x), (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.3, (100,255,155), 1, cv2.LINE_AA)
            #draw the y coordinate in the frame
            cv2.putText(frame, "y:" + format(y), (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.3, (100,255,155), 1, cv2.LINE_AA) 
        IPython.display.display(PIL.Image.fromarray(frame))
        IPython.display.clear_output(wait=True)
        
finally:
    camera.close()
```
